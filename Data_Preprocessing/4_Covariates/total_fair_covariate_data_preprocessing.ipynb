```python
unique_values_group = list(dd_ts_fresh['group'].unique())
print(len(unique_values_group), "\n")
print(unique_values_group)
```

```python
zones_time_idx = {}
for i in unique_values_group:
    df = dd_ts_fresh[dd_ts_fresh['group'] == i]
    time_idx_list = df['time_idx'].tolist()
    first_idx = time_idx_list[0]
    last_idx = time_idx_list[-1]
    zones_time_idx[i] = [first_idx, last_idx]
    print("Zone ", str(i), " has ", df.shape[0], " number of rows in df_ts_fresh \n")
    print("First idx: ", first_idx, "; Last idx: ", last_idx, "\n")
```

```python
import numpy as np

# Assuming total_fair_ts_final is already defined
total_fair_ts_final = pd.DataFrame(columns=['total_fair'])

# Your existing code
for zone, indexes in zones_time_idx.items():
    index_list = list(np.arange(indexes[0], indexes[1] + 1))
    df = total_fair_ts[[str(zone)]]
    df2 = df[df.index.isin(index_list)]

    # Cumulatively add rows to total_fair_ts_final
    total_fair_ts_final = pd.concat([total_fair_ts_final, df2.rename(columns={str(zone): 'total_fair'})])
```
